```python
import numpy as np
def f(a,b):
  if len(a[1]) == len(b.T[1]): 
    return np.dot(a,b).T.astype(np.int64)
  else:
    return 0
a = np.array(([4.0,92.0,6.0],[-15.0,6.0,10.0],[9.0,8.0,3.0]))
b = np.array(([14.0,7.0,7.0],[6.0,40.0,13.0],[65.0,3.0,7.0]))
print(f(a,b))
```

```python
import numpy as np
def f(x1,x2):
  return x1**2 + 2*x2**2 + np.exp(x1 + x2)
x1 = np.random.randint(-1000,1000)
x2 = np.random.randint(-1000,1000)
e = 0.5
dydx1 = np.gradient(f, x1)
dydx2 = np.gradient(f, x2)
F = np.array([dydx1,dydx2])
while (abs(x1) >= e) and (abs(x2) >= e):
  x1 = x1 - e*F[0]
  x2 = x2 - e*F[1]
print(x1,x2)
```
